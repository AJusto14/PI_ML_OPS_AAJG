Como sabemos, cada punto del gráfico es una película y la posición del punto indica la duración de la película y la recompensa obtenida. Pero esto no nos dice mucho, entonces podemos intentar hacer un análisis un poco más allá y tener más a produnfidad de estos datos.

```python
#Importamos otra librería
import statsmodels.api as sm
```

```python
#Creamos un nuevo DataFrame con las columnas 'revenue' y 'runtime' sin valores nulos
#Esto para que tengamos un mejor resultado
df_cleaned = parte_6[['revenue', 'runtime']].dropna()

#Aquí, realizamos un ajuste de regresión lineal
#Usamos la librería sm y la función add_constant para agregar una
#constante a una variable en el anñalisis de regresión para hacer
#una matríz bidimensional entre una columna de unos y 'runtime'
#preparando el dataset para el análisis de regresión lineal
X = sm.add_constant(df_cleaned['runtime'])
#Asignamos a otra variable revenue sin nulos o NA
y = df_cleaned['revenue']
#Con esto creamos un objeto de modelo de regresión lineal de mínimos
#cuadrados ordinarios (OLS) para ajustar y estimar el modelo lineal
#a los datos
model = sm.OLS(y, X)
#Entrenamos al modelo 
results = model.fit()

#Obtenemos los parámetros de la regresión lineal
intercept = results.params[0]
slope = results.params[1]

# Crear el gráfico de dispersión con la línea de regresión
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='runtime', y='revenue')
plt.plot(df_cleaned['runtime'], intercept + slope * df_cleaned['runtime'], color='red', linewidth=2)
plt.xlabel('Duración de la película (minutos)')
plt.ylabel('Recompensa obtenida')
plt.title('Relación entre Duración y Recompensa con Ajuste de Regresión')

plt.show()
```

En conclusión de este gráfico podemos ver que no se tiene una buena correlación, porque si bien existe una película con mayor recompensa pero con menos de 200 minutos, y si vemos las películas de muchísima duración no tienen un alto índice de recomenpensa.

Por último, tomando el dataset completo de Movies, veamos si de acuerdo a su presupuesto existe alguna correlación con su popularidad, es decir, queremos conocer si mientras más presupuesto, mejor la película.

```python
#Primero nos aseguramos que los datos tengan el formato correcto
#es decir, que por ser números, sean de tipo numérico
movies['budget'] = pd.to_numeric(movies['budget'], errors='coerce')
movies['popularity'] = pd.to_numeric(movies['popularity'], errors='coerce')
    #nota que tenemos un parámetro: errors, lo que hace es que convierte
    #los valores no numéricos en valores NaN

#Ahora calculamos la correlación entre las columnas ya acomodadas
correlation = movies['budget'].corr(movies['popularity'])
print("Correlation entre Budget y Popularidad es:", correlation)

#Creamos el gráfico
#sns.scatterplot(data=movies, x='budget', y='popularity')
sns.regplot(data=movies, x='budget', y='popularity', line_kws={'color': 'red'})
    #Cambiamos a regplot para ver la línea de tendencia, esto
    #mediante una línea de regresión en los puntos
plt.xlabel('Budget')
plt.ylabel('Popularity')
plt.title('Correlation entre Budget y Popularity')

plt.show()
```

Como vemos, el coeficiente que nos da el texto es cercano a 0, entonces concluimos que tenemos una correlación inexistente o debil, adicional, viendo la línea de tendencia podemos ver que no es importante el presupuesto, porque si bien hay unas con mucho presupuesto pero con baja popularidad y el outlier indica la más alta popularidad pero con poco presupuesto.

```python
#Ya para finalizar con este datset, se me ocurrió que pudieramos
#ver si hay alguna correlación entre el día de estreno y su
#recaudación

#Hacemos lo mismo que antes con revenue
movies['revenue'] = pd.to_numeric(movies['revenue'], errors='coerce')

#Ahora convertimos en formato numérico los meses en lugar de 
#string que está actualmente para que sea más sencilla la comparación
#para esto usaremos map y un diccionario con los meses
month_mapping = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6, 'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
movies['release_month'] = movies['release_month'].map(month_mapping)

#Ahora calculamos la correlación entre las columnas ya acomodadas
correlation = movies['revenue'].corr(movies['release_month'])
print("Correlation entre Revenue y Release Month:", correlation)

#Ahora lo vemos graficamente
plt.scatter(movies['release_month'], movies['revenue'])
plt.xlabel('Release Month')
plt.ylabel('Revenue')
plt.title('Correlation entre Revenue y Release Month')

plt.show()
```

Como podemos ver en el resultado dicho antes de la gráfica, la correlación es muy cercana a 0, entonces concluimos que no existe correlación entre estas dos variables.

```python
#Aquí podemos ver cual es el director con más películas

#Primero, reordenamos el dataset para ver solo los directores
directores = crew.loc[crew['job'] == 'Director']

#Ahora contamos las veces que aparece cada nombre
director_conteo = directores['name'].value_counts()

#Ahora encontramos el nombre que tiene más películas
director_frecuencia = director_conteo.idxmax()

#Ahora obtenemos el número de veces que aparece el nombre del
#director
conteo_frecuente = director_conteo.max()

#vemos el resultado
print("El director con más películas:", director_frecuencia)
print("Número de Películas:", conteo_frecuente)

#Y si lo graficamos se ve así, lo limitamos a solo 5 nombres
#los que tienen más películas
director_conteo[:5].plot(kind='bar')
plt.xlabel('Director')
plt.ylabel('Veces')
plt.title('Número de películas por Director')

plt.show()
```
